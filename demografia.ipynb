```python
def plot_it(anno=1982, min_lav=20, max_lav=60, colormap='Reds'):

    fig, ax = pl.subplots(1, 1, figsize=(2, 1))
    
    try:
        i = np.argwhere(data.index==anno)[0][0]
    except IndexError:
        i = np.argwhere(data.index==2018)[0][0]
    n_steps = 5
    
    cmap = pl.cm.__getattribute__(colormap)

    [ax.step(range(len(data.iloc[i-s])), data.iloc[i-s]*1e-6, color=c, zorder=0)
     for s, c in zip(range(n_steps-1, -1, -1), cmap(np.linspace(0.1, 0.9, n_steps))) if i-s>=0]
    ax.set_title('Anno %d'%data.index[i])
    
    grouped = [np.sum([data.iloc[i][a]*1e-6 for a in range(min_lav)]),
               np.sum([data.iloc[i][a]*1e-6for a in range(min_lav, max_lav)]),
               np.sum([np.nan_to_num(data.iloc[i][a])*1e-6 for a in range(max_lav, 110)]),
               ]
    ax.text(110, 0.8, 'Giovani: %.1f'%grouped[0],
            color='b', )
    ax.text(110, 0.6, 'Lavoratori: %.1f'%grouped[1],
            color='g', fontweight='bold')
    ax.text(110, 0.4, 'Pensionati: %.1f'%grouped[2],
            color='r', )
    ax.text(110, 0.2, 'Totale: %.1f'%np.sum(data.iloc[i]*1e-6))

    ax.set_xlim(0, 110)
    ax.set_ylim(0, 1.2)
    ax.set_yticks((0, 0.5, 1))
    ax.set_ylabel('mln di persone')
    ax.set_xlabel('Età')
    ax.fill_between([min_lav, max_lav], [1.2, 1.2], color=(0.8, 1, 0.8), zorder=0)
    
interact(plot_it, anno=(1982, 2050, 1), colormap=dir(pl.cm))
```
